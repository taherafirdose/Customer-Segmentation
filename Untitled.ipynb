```python
# Assuming 'Sex', 'Marital status', 'Education', 'Occupation', 'Settlement size' are categorical
df.drop('Income Category', axis=1, inplace=True)

```

```python
# Create transformers for categorical and numerical data
categorical_transformer = OneHotEncoder()
numerical_transformer = Pipeline(steps=[('scaler', StandardScaler())])

# Combine transformers into a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, num_columns),
        ('cat', categorical_transformer, cat_columns)
    ])

# Apply the transformations to your DataFrame
prepared_data = preprocessor.fit_transform(df)
```

```python
# Elbow Method
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(prepared_data)
    sse.append(kmeans.inertia_)
   

plt.plot(range(1, 11), sse)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
# Replace 'n_clusters' with the chosen number of clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(prepared_data)
```

```python
df['clusters']=clusters
```

```python
# Number of plots per row
plots_per_row = 2

# Calculate the total number of rows needed
num_rows = len(cat_columns) // plots_per_row + (len(cat_columns) % plots_per_row > 0)

# Create a figure with the required number of subplots
plt.figure(figsize=(plots_per_row * 8, num_rows * 5))  # Adjust the figure size as needed

for i, cat_col in enumerate(cat_columns):
    plt.subplot(num_rows, plots_per_row, i + 1)  # Creating subplots
    cluster_distribution = pd.crosstab(df['clusters'], df[cat_col])
    cluster_proportions = cluster_distribution.div(cluster_distribution.sum(axis=1), axis=0)
    cluster_proportions.plot(kind='bar', stacked=False, ax=plt.gca())  # Plot on the current axes

    plt.title(f'Distribution of {cat_col} by Cluster')
    plt.ylabel('Proportion')
    plt.xlabel('Cluster')

plt.tight_layout()
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Assuming your data is stored in 'scaled_df'
# Generate the linkage matrix
Z = linkage(prepared_data, method='ward')

# Plot the dendrogram
plt.figure(figsize=(12, 8))
dendrogram(Z, truncate_mode='lastp', p=12, leaf_rotation=45., leaf_font_size=15., show_contracted=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.show()

# You can also apply Agglomerative Clustering to obtain the cluster labels
# Choose the number of clusters based on the dendrogram
agglom = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
clusters = agglom.fit_predict(prepared_data)

# Add clusters to your DataFrame
df['Cluster'] = clusters
```
